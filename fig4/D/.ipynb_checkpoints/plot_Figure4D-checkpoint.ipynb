```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from obspy.signal.filter import envelope
from obspy.geodetics import gps2dist_azimuth
from obspy.signal import util
from obspy.signal import cross_correlation
import numpy as np
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
from obspy import Trace, read
from obspy.signal.util import smooth
from obspy.signal.tf_misfit import cwt
from datetime import timedelta, datetime
import ScientificColourMaps6 as SCM6
import matplotlib.dates as mdates
from scipy.fft import fftfreq 
from scipy.signal import butter, filtfilt
from obspy.signal.util import smooth
from pytides.tide import Tide
import matplotlib.patches as patches
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import matplotlib as mpl
from matplotlib.lines import Line2D
```

```python
mpl.rc('font', family='sans-serif')
mpl.rc('font', serif='Arial')
mpl.rc('font', size=11)
```

```python
def seiche_sim_to_trace_resample(in_fn):
    # Function to make an ObsPy trace object from a tsunami time series csv file
    df = pd.read_csv(in_fn, header=None, names=["t", "a"], delim_whitespace=True)
    fs = 1000
    seiche_offset = 0
    df = df.set_index(pd.to_datetime(df['t'], unit='s'), drop=False)
    resample_index = pd.date_range(start=df.index[0], end=df.index[-1], freq=f'{fs}ms')
    dummy_frame = pd.DataFrame(np.NaN, index=resample_index, columns=df.columns)
    df_interp = df.combine_first(dummy_frame).interpolate('time').resample(f'{fs}ms').fillna("pad")
    tr = Trace(data=np.array(df_interp["a"].values))
    tr.stats.delta = fs/1000
    tr.stats.starttime = UTCDateTime(df_interp.iloc[0].t) 
    tr2 = tr.copy()
    tr2.detrend()
    tr2.detrend(type="demean")
    tr2.detrend('linear')
    return tr2
```

```python
t_ev = UTCDateTime(2023, 9, 16, 12, 35)  # Origin time of landslide

# Time window of interest
t1 = UTCDateTime(2023, 9,  15,  0, 0, 1)
t2 = UTCDateTime(2023, 9, 20, 0, 0, 1)
```

```python
# Get trace of simulated tsunami time series
tr2 = seiche_sim_to_trace_resample("../../seiche/eta1-Gauge21.txt")
tr2.stats.starttime = t_ev
yy = smooth(envelope(tr2.data), 200)
```

```python
# Read in observed tide gauage data and de-tide
df = pd.read_csv("DicksonFjord_CTD_18102023.csv")  # This file is from https://doi.org/10.14284/637
df["dt"] = pd.to_datetime(df["Timestamp"])
df.index = df["dt"]

df_reind = df.reindex(pd.date_range(start=df.index.min(), end=df.index.max(), freq='15Min'))    
df_reind.interpolate(method='linear', inplace=True)   

t = df_reind["dt"].values.astype('datetime64[s]').tolist()
heights = df_reind["Depth_m"].values

df_reind2 = df_reind[df_reind["dt"].between(t1.datetime, t2.datetime)]

prediction_t0 = t1.datetime
hours = df_reind2.dt.values.astype('datetime64[s]').tolist()
times = Tide._times(prediction_t0, hours)
my_tide = Tide.decompose(heights, t)
my_prediction = my_tide.at(times)

tr = Trace(data=np.array(df_reind2["Depth_m"]-my_prediction))
tr.stats.starttime = UTCDateTime(df_reind2.iloc[0]["dt"])
tr.stats.delta = 15*60
tr.detrend("linear")
tr.detrend("demean")

tr_CTD_raw = Trace(data=np.array(df_reind2["Depth_m"]))
tr_CTD_raw.stats.starttime = UTCDateTime(df_reind2.iloc[0]["dt"])
tr_CTD_raw.stats.delta = 15*60
```

```python
# Now do the plotting
fig = plt.figure(figsize=(13, 5))
ax = plt.subplot()
ax.semilogy(tr2.times(reftime=t_ev)/60/60, yy, label="Smoothed envelope of\nsimulated CTD signal", alpha=0.8, c="darkgreen") #, marker="*")
ax.axhline(0.03, c="blue", ls="--", label="Average pre-event\nCTD de-tided background\nnoise level", alpha=0.8)
ax.semilogy(tr.times(reftime=t_ev)/60/60, envelope(tr.data), label="Envelope of de-tided\nobserved CTD gauge", marker="o", mew=0.2, mec="k", alpha=0.8, ms=4, c="red")

ax.set_xlabel("Time since landslide (2023-09-16 12:35:00) in hours")
ax.set_ylabel("Water height (m)")
ax.set_xlim(-30, 15)
ax.set_ylim([0.01, 4])
ax.grid(ls="--", alpha=0.5)

inset_ax = inset_axes(ax,
                    width="50%", # width = 30% of parent_bbox
                    height=1.4, # height : 1 inch
                    loc=1)

inset_ax.axvspan(-7, 15, fc="lightgray")
inset_ax.plot(tr_CTD_raw.times(reftime=t_ev)/60/60, tr_CTD_raw.data, c="darkorange", lw=1.2)
inset_ax.grid(ls="--", alpha=0.5)
inset_ax.set_xlabel("Time since landslide (2023-09-16 12:35:00) in hours")
inset_ax.set_ylabel("Water height (m)")


handles, labels = ax.get_legend_handles_labels()
handles.append(Line2D([0], [0], label='manual line', color='darkorange'))
labels.append("Raw observed\nCTD gauge data")

ax.legend(handles, labels, loc=2)

plt.savefig("Figure4D.png", dpi=300, transparent=True, bbox_inches="tight")
```
