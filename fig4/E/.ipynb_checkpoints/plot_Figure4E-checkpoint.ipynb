```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from obspy.signal.filter import envelope
from obspy.geodetics import gps2dist_azimuth
from obspy.signal import util
from obspy.signal import cross_correlation
import numpy as np
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
from obspy import Trace, read
from obspy.signal.util import smooth
from obspy.signal.tf_misfit import cwt
from datetime import timedelta, datetime
import ScientificColourMaps6 as SCM6
import matplotlib.dates as mdates
from scipy.fft import fftfreq 
from scipy.signal import butter, filtfilt
from obspy.signal.util import smooth
from pytides.tide import Tide
import matplotlib.patches as patches
```

```python
# Read in time series
df = pd.read_csv("../../seiche/Gauge0-Manning0-0x3-dx=3-18hrs.txt",
                 header=None, names=["t", "a"], delim_whitespace=True)
fs = 50  # Interpolate to this sample frequency (in ms)
seiche_offset = 0
df = df.set_index(pd.to_datetime(df['t'], unit='s'), drop=False)
resample_index = pd.date_range(start=df.index[0], end=df.index[-1], freq=f'{fs}ms')
dummy_frame = pd.DataFrame(np.NaN, index=resample_index, columns=df.columns)
df_interp = df.combine_first(dummy_frame).interpolate('time').resample(f'{fs}ms').fillna("pad")
#df_interp = df_interp[df_interp.t.between(seiche_offset, 58300)]
print(df_interp)
tr = Trace(data=np.array(df_interp["a"].values))
tr.stats.delta = fs/1000
print(tr.stats.delta)
tr.stats.starttime = UTCDateTime(df_interp.iloc[0].t) 
print(tr.stats.starttime)
tr2 = tr.copy()
print(tr2.data[-1])
tr2.detrend()
tr2.detrend(type="demean")
tr2.detrend('linear')
FILTER_MIN_FREQ_IN_HZ = 0.009
FILTER_MAX_FREQ_IN_HZ = 0.013
tr2_unfilt = tr2.copy()
tr2.filter(
    'bandpass', freqmin=FILTER_MIN_FREQ_IN_HZ, freqmax=FILTER_MAX_FREQ_IN_HZ)

fig = plt.figure(figsize=(8, 5))
ax = plt.subplot()

ax.plot(tr2.times(reftime=tr2.stats.starttime)/3600, tr2.data, lw=0.2, c="darkgreen", label="dx 3m; Manning = 0.03; bandpass filtered 9-13 mHz")


ax.set_ylabel("Water surface elevation (m)", fontsize=13)
ax.set_xlabel("Time since landslide (hours)", fontsize=13)
ax.grid(ls="--", alpha=0.5)
ax.margins(x=0.02)

f = 1/87
wo = 2 * np.pi *f
Q_all = [500, 3000]
t_all = np.linspace(0, 16.2*3600, 100)
colors = ["magenta", "blue"]
for nq, Q in enumerate(Q_all):
    if Q == 500:
        Ao = np.max(tr2.data) * 0.8
    else:
        Ao = np.max(tr2.data) * 0.5
    x = Ao * np.exp(-wo*(t_all)/(2*Q))
    ax.plot((t_all)/3600, x, label=f"$Q$ = {Q}", ls="--", alpha=0.7, lw=2.0, c=colors[nq])
ax.legend()
plt.savefig("Figure4E.png", dpi=300, bbox_inches="tight")
```
